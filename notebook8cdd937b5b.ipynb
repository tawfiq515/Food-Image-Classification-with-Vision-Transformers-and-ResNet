```python
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Subset, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import time
from transformers import ViTForImageClassification, ViTImageProcessor
import os
from torch.optim.lr_scheduler import StepLR, CosineAnnealingLR
from PIL import Image
import warnings
warnings.filterwarnings('ignore')


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")


# Create data directory for storing dataset
data_dir = "./food101"
if not os.path.exists(data_dir):
    os.makedirs(data_dir)
    print(f"Created directory: {data_dir}")


def prepare_food101_data(use_subset=True, subset_classes=20):
    """
    Prepare Food-101 dataset with train/val/test splits
    Args:
        use_subset: Whether to use a subset of classes for faster training
        subset_classes: Number of classes to use if use_subset=True
    """
    # Data augmentation and normalization for ViT
    train_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(degrees=15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.RandomApply([transforms.GaussianBlur(3)], p=0.2),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    print("Loading Food-101 dataset...")
    print("This may take several minutes on first download (5GB+)...")

    try:
        # Download and load the full dataset
        train_dataset_full = datasets.Food101(
            root=data_dir,
            split='train',
            download=True,
            transform=train_transform
        )

        test_dataset_full = datasets.Food101(
            root=data_dir,
            split='test',
            download=True,
            transform=val_transform
        )

        print(f"Dataset loaded successfully!")
        print(f"Total training samples: {len(train_dataset_full)}")
        print(f"Total test samples: {len(test_dataset_full)}")
        print(f"Number of classes: {len(train_dataset_full.classes)}")

        # Use subset of classes for faster experimentation
        if use_subset:
            print(f"Using subset of {subset_classes} classes for faster training...")

            # Get class indices for subset
            selected_classes = list(range(subset_classes))

            # Filter training data
            train_indices = [i for i, (_, label) in enumerate(train_dataset_full)
                           if label in selected_classes]
            train_dataset = Subset(train_dataset_full, train_indices)

            # Filter test data
            test_indices = [i for i, (_, label) in enumerate(test_dataset_full)
                          if label in selected_classes]
            test_dataset = Subset(test_dataset_full, test_indices)

            num_classes = subset_classes
            class_names = [train_dataset_full.classes[i] for i in selected_classes]
        else:
            train_dataset = train_dataset_full
            test_dataset = test_dataset_full
            num_classes = len(train_dataset_full.classes)
            class_names = train_dataset_full.classes

        # Split training data into train and validation
        train_size = int(0.85 * len(train_dataset))
        val_size = len(train_dataset) - train_size

        train_subset, val_dataset = random_split(
            train_dataset,
            [train_size, val_size],
            generator=torch.Generator().manual_seed(42)
        )

        print(f"\nUsing {num_classes} food classes:")
        print(f"Selected classes: {', '.join(class_names[:10])}{'...' if len(class_names) > 10 else ''}")

    except Exception as e:
        print(f"Error loading Food-101 dataset: {e}")
        print("Creating synthetic food dataset for testing...")

        class SyntheticFoodDataset(torch.utils.data.Dataset):
            def __init__(self, size=2000, img_size=224, num_classes=20):
                self.size = size
                self.img_size = img_size
                self.num_classes = num_classes
                self.classes = [f"food_class_{i}" for i in range(num_classes)]

            def __len__(self):
                return self.size

            def __getitem__(self, idx):
                # Generate synthetic food-like image
                img = torch.rand(3, self.img_size, self.img_size)
                # Add some texture patterns
                img[:, idx % self.img_size::20, :] += 0.3
                img[:, :, idx % self.img_size::15] += 0.3
                img = torch.clamp(img, 0, 1)

                label = idx % self.num_classes
                return img, label

        # Create synthetic datasets
        full_dataset = SyntheticFoodDataset(size=2000, num_classes=subset_classes if use_subset else 101)
        train_subset = Subset(full_dataset, range(1200))
        val_dataset = Subset(full_dataset, range(1200, 1600))
        test_dataset = Subset(full_dataset, range(1600, 2000))
        num_classes = subset_classes if use_subset else 101
        class_names = [f"synthetic_food_{i}" for i in range(num_classes)]

    # Create data loaders with INCREASED BATCH SIZE
    batch_size = 64  # Increased from 32 for faster training
    train_loader = DataLoader(
        train_subset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=4,
        pin_memory=True if torch.cuda.is_available() else False,
        drop_last=True
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=4,
        pin_memory=True if torch.cuda.is_available() else False
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=4,
        pin_memory=True if torch.cuda.is_available() else False
    )

    print(f"\nData split:")
    print(f"Training samples: {len(train_subset)}")
    print(f"Validation samples: {len(val_dataset)}")
    print(f"Test samples: {len(test_dataset)}")
    print(f"Training batches: {len(train_loader)}")
    print(f"Validation batches: {len(val_loader)}")
    print(f"Test batches: {len(test_loader)}")

    return train_loader, val_loader, test_loader, num_classes, class_names


def setup_vit_model(num_classes, model_size="base", freeze_backbone=False):
    """
    Setup Vision Transformer model with different fine-tuning strategies
    Args:
        num_classes: Number of target classes
        model_size: "base" or "large"
        freeze_backbone: Whether to freeze the backbone and only train classifier
    """
    print(f"\nLoading pretrained ViT-{model_size} model...")

    if model_size == "base":
        model_name = "google/vit-base-patch16-224"
    elif model_size == "large":
        model_name = "google/vit-large-patch16-224"
    else:
        model_name = "google/vit-base-patch16-224"

    print(f"Model: {model_name}")

    model = ViTForImageClassification.from_pretrained(
        model_name,
        num_labels=num_classes,
        ignore_mismatched_sizes=True,
        image_size=224
    )

    # Implement different fine-tuning strategies
    if freeze_backbone:
        print("Freezing backbone layers, training only classifier head...")
        # Freeze all ViT layers except classifier
        for name, param in model.named_parameters():
            if "classifier" not in name:
                param.requires_grad = False

        strategy = "frozen_backbone"
    else:
        print("Fine-tuning all layers...")
        # Unfreeze all parameters
        for param in model.parameters():
            param.requires_grad = True

        strategy = "full_finetune"

    # Print model info
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    print(f"Frozen parameters: {total_params - trainable_params:,}")

    model = model.to(device)
    return model, strategy


def setup_resnet_model(num_classes, freeze_backbone=False):
    """Setup ResNet50 model for comparison"""
    print(f"\nLoading pretrained ResNet50 model...")

    model = torchvision.models.resnet50(weights='IMAGENET1K_V2')

    if freeze_backbone:
        print("Freezing backbone layers, training only classifier head...")
        # Freeze all layers except final classifier
        for name, param in model.named_parameters():
            if "fc" not in name:
                param.requires_grad = False
        strategy = "frozen_backbone"
    else:
        print("Fine-tuning all layers...")
        for param in model.parameters():
            param.requires_grad = True
        strategy = "full_finetune"

    # Replace the final fully connected layer
    num_ftrs = model.fc.in_features
    model.fc = nn.Sequential(
        nn.Dropout(0.3),
        nn.Linear(num_ftrs, num_classes)
    )

    # Print model info
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    print(f"Frozen parameters: {total_params - trainable_params:,}")

    model = model.to(device)
    return model, strategy


def train_model(model, train_loader, val_loader, num_epochs=8, model_name="ViT", strategy="full_finetune"):
    """Train the model with advanced fine-tuning techniques"""
    print(f"\nTraining {model_name} ({strategy}) for {num_epochs} epochs...")

    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

    # Different optimization strategies based on fine-tuning approach with NORMAL LEARNING RATES
    if strategy == "frozen_backbone":
        # Higher learning rate for classifier-only training
        optimizer = optim.AdamW(
            model.parameters(),
            lr=0.002,  # Normal learning rate instead of 1e-3
            weight_decay=0.01
        )
        scheduler = StepLR(optimizer, step_size=3, gamma=0.5)  # Adjusted for 8 epochs
    else:
        # Lower learning rate for full fine-tuning
        if "ViT" in model_name:
            optimizer = optim.AdamW(
                model.parameters(),
                lr=0.001,  # Normal learning rate instead of 3e-4
                weight_decay=0.01,
                betas=(0.9, 0.999)
            )
        else:  # ResNet
            optimizer = optim.AdamW(
                model.parameters(),
                lr=0.001,  # Normal learning rate instead of 1e-4
                weight_decay=1e-4
            )

        scheduler = CosineAnnealingLR(optimizer, T_max=num_epochs, eta_min=1e-6)

    # Initialize metrics storage
    train_losses, val_losses = [], []
    train_accs, val_accs = [], []
    best_val_acc = 0.0

    for epoch in range(num_epochs):
        start_time = time.time()

        # Training phase
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for batch_idx, (inputs, labels) in enumerate(train_loader):
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            # Forward pass
            if "ViT" in model_name:
                outputs = model(inputs).logits
            else:
                outputs = model(inputs)

            loss = criterion(outputs, labels)

            # Backward pass
            loss.backward()

            # Gradient clipping for stability
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()

            # Statistics
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

            # Print progress
            if batch_idx % 25 == 0 and batch_idx > 0:  # Reduced frequency for batch size 64
                current_lr = optimizer.param_groups[0]['lr']
                print(f'  Batch [{batch_idx}/{len(train_loader)}] Loss: {loss.item():.4f} LR: {current_lr:.6f}', end='\r')

        train_loss = running_loss / len(train_loader)
        train_acc = 100.0 * correct / total
        train_losses.append(train_loss)
        train_accs.append(train_acc)

        # Validation phase
        model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)

                if "ViT" in model_name:
                    outputs = model(inputs).logits
                else:
                    outputs = model(inputs)

                loss = criterion(outputs, labels)
                running_loss += loss.item()

                _, predicted = outputs.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

        val_loss = running_loss / len(val_loader)
        val_acc = 100.0 * correct / total
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), f'best_{model_name.lower()}_{strategy}_food101.pth')

        scheduler.step()

        epoch_time = time.time() - start_time
        current_lr = optimizer.param_groups[0]['lr']
        print(f"Epoch {epoch+1}/{num_epochs} - Time: {epoch_time:.2f}s - LR: {current_lr:.6f}")
        print(f"  Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.2f}%")
        print(f"  Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.2f}%")

    print(f"\nBest validation accuracy: {best_val_acc:.2f}%")
    return (train_losses, val_losses, train_accs, val_accs), best_val_acc


def evaluate_model(model, test_loader, model_name="ViT"):
    """Evaluate model on test set with detailed metrics"""
    model.eval()
    correct = 0
    total = 0
    class_correct = {}
    class_total = {}

    print(f"Evaluating {model_name} on test set...")

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            if "ViT" in model_name:
                outputs = model(inputs).logits
            else:
                outputs = model(inputs)

            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

            # Per-class accuracy
            for i in range(labels.size(0)):
                label = labels[i].item()
                class_total[label] = class_total.get(label, 0) + 1
                if predicted[i] == labels[i]:
                    class_correct[label] = class_correct.get(label, 0) + 1

    accuracy = 100.0 * correct / total
    print(f"{model_name} Test Accuracy: {accuracy:.2f}%")

    # Print per-class accuracy for first few classes
    print(f"Per-class accuracy (first 5 classes):")
    for class_id in sorted(list(class_total.keys())[:5]):
        if class_id in class_correct:
            class_acc = 100.0 * class_correct[class_id] / class_total[class_id]
            print(f"  Class {class_id}: {class_acc:.1f}%")

    return accuracy


def plot_comparison_advanced(results_dict):
    """Plot advanced training metrics comparison"""
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    colors = {'ViT_full_finetune': 'red', 'ViT_frozen_backbone': 'orange',
              'ResNet50_full_finetune': 'blue', 'ResNet50_frozen_backbone': 'cyan'}

    for model_key, (metrics, _) in results_dict.items():
        epochs = range(1, len(metrics[0]) + 1)
        color = colors.get(model_key, 'black')
        label = model_key.replace('_', ' ').title()

        # Training Loss
        axes[0, 0].plot(epochs, metrics[0], color=color, label=label, linewidth=2)

        # Validation Loss
        axes[0, 1].plot(epochs, metrics[1], color=color, label=label, linewidth=2)

        # Training Accuracy
        axes[1, 0].plot(epochs, metrics[2], color=color, label=label, linewidth=2)

        # Validation Accuracy
        axes[1, 1].plot(epochs, metrics[3], color=color, label=label, linewidth=2)

    # Formatting
    titles = ["Training Loss", "Validation Loss", "Training Accuracy (%)", "Validation Accuracy (%)"]
    for i, ax in enumerate(axes.flat):
        ax.set_title(titles[i], fontsize=14, fontweight='bold')
        ax.set_xlabel("Epochs")
        ax.legend()
        ax.grid(True, alpha=0.3)
        if i >= 2:  # Accuracy plots
            ax.set_ylabel("Accuracy (%)")
        else:  # Loss plots
            ax.set_ylabel("Loss")

    plt.suptitle("Fine-tuning Strategies Comparison on Food-101 (8 Epochs)", fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig('food101_finetuning_comparison_8epochs.png', dpi=300, bbox_inches='tight')
    plt.show()


def visualize_predictions(model, loader, class_names, model_name="ViT", num_images=8):
    """Visualize model predictions on food images"""
    model.eval()

    images_shown = 0
    fig, axes = plt.subplots(2, num_images//2, figsize=(20, 10))
    axes = axes.ravel()

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)

            if "ViT" in model_name:
                outputs = model(inputs).logits
            else:
                outputs = model(inputs)

            probabilities = torch.nn.functional.softmax(outputs, dim=1)
            _, preds = torch.max(outputs, 1)

            for i in range(inputs.size(0)):
                if images_shown >= num_images:
                    break

                img = inputs[i].cpu()
                # Denormalize
                img = img * torch.tensor([0.229, 0.224, 0.225]).view(3,1,1)
                img = img + torch.tensor([0.485, 0.456, 0.406]).view(3,1,1)
                img = torch.clamp(img, 0, 1)

                axes[images_shown].imshow(np.transpose(img.numpy(), (1, 2, 0)))

                true_label = labels[i].item()
                pred_label = preds[i].item()
                confidence = probabilities[i][pred_label].item()

                # Get class names if available
                true_name = class_names[true_label] if true_label < len(class_names) else f"Class {true_label}"
                pred_name = class_names[pred_label] if pred_label < len(class_names) else f"Class {pred_label}"

                color = 'green' if true_label == pred_label else 'red'
                axes[images_shown].set_title(
                    f"True: {true_name}\nPred: {pred_name}\nConf: {confidence:.2f}",
                    fontsize=10, color=color, fontweight='bold'
                )
                axes[images_shown].axis("off")

                images_shown += 1

            if images_shown >= num_images:
                break

    plt.suptitle(f"{model_name} Predictions on Food-101", fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{model_name.lower()}_predictions_food101_8epochs.png', dpi=300, bbox_inches='tight')
    plt.show()


def main():
    """Main execution function with comprehensive fine-tuning experiments"""
    print("="*80)
    print("Fine-tuning Pre-trained Vision Transformers on Food-101")
    print("Comparing Different Fine-tuning Strategies (8 Epochs, Batch Size 64)")
    print("="*80)

    # Prepare data - using subset for faster experimentation
    train_loader, val_loader, test_loader, num_classes, class_names = prepare_food101_data(
        use_subset=True, subset_classes=20
    )

    # Store all results
    results = {}
    test_accuracies = {}

    # Experiment 1: ViT with full fine-tuning
    print("\n" + "="*80)
    print("EXPERIMENT 1: ViT Base - Full Fine-tuning")
    print("="*80)
    vit_model_full, strategy = setup_vit_model(num_classes, model_size="base", freeze_backbone=False)
    vit_metrics_full, vit_best_acc_full = train_model(
        vit_model_full, train_loader, val_loader,
        num_epochs=8, model_name="ViT", strategy=strategy
    )
    results["ViT_full_finetune"] = (vit_metrics_full, vit_best_acc_full)

    # Experiment 2: ViT with frozen backbone
    print("\n" + "="*80)
    print("EXPERIMENT 2: ViT Base - Frozen Backbone")
    print("="*80)
    vit_model_frozen, strategy = setup_vit_model(num_classes, model_size="base", freeze_backbone=True)
    vit_metrics_frozen, vit_best_acc_frozen = train_model(
        vit_model_frozen, train_loader, val_loader,
        num_epochs=8, model_name="ViT", strategy=strategy
    )
    results["ViT_frozen_backbone"] = (vit_metrics_frozen, vit_best_acc_frozen)

    # Experiment 3: ResNet50 with full fine-tuning
    print("\n" + "="*80)
    print("EXPERIMENT 3: ResNet50 - Full Fine-tuning")
    print("="*80)
    resnet_model_full, strategy = setup_resnet_model(num_classes, freeze_backbone=False)
    resnet_metrics_full, resnet_best_acc_full = train_model(
        resnet_model_full, train_loader, val_loader,
        num_epochs=8, model_name="ResNet50", strategy=strategy
    )
    results["ResNet50_full_finetune"] = (resnet_metrics_full, resnet_best_acc_full)

    # Experiment 4: ResNet50 with frozen backbone
    print("\n" + "="*80)
    print("EXPERIMENT 4: ResNet50 - Frozen Backbone")
    print("="*80)
    resnet_model_frozen, strategy = setup_resnet_model(num_classes, freeze_backbone=True)
    resnet_metrics_frozen, resnet_best_acc_frozen = train_model(
        resnet_model_frozen, train_loader, val_loader,
        num_epochs=8, model_name="ResNet50", strategy=strategy
    )
    results["ResNet50_frozen_backbone"] = (resnet_metrics_frozen, resnet_best_acc_frozen)

    # Plot comprehensive comparison
    print("\n" + "="*80)
    print("PLOTTING COMPREHENSIVE COMPARISON")
    print("="*80)
    plot_comparison_advanced(results)

    # Final evaluation on test set
    print("\n" + "="*80)
    print("FINAL EVALUATION ON TEST SET")
    print("="*80)

    models_to_evaluate = [
        (vit_model_full, "ViT Full Fine-tune"),
        (vit_model_frozen, "ViT Frozen Backbone"),
        (resnet_model_full, "ResNet50 Full Fine-tune"),
        (resnet_model_frozen, "ResNet50 Frozen Backbone")
    ]

    for model, name in models_to_evaluate:
        test_acc = evaluate_model(model, test_loader, name)
        test_accuracies[name] = test_acc

    # Visualize predictions from best model
    best_model_name = max(test_accuracies, key=test_accuracies.get)
    best_model = next(model for model, name in models_to_evaluate if name == best_model_name)

    print(f"\n" + "="*80)
    print(f"VISUALIZING PREDICTIONS FROM BEST MODEL: {best_model_name}")
    print("="*80)
    visualize_predictions(best_model, test_loader, class_names, best_model_name, num_images=8)

    # Final comprehensive summary
    print("\n" + "="*80)
    print("COMPREHENSIVE RESULTS SUMMARY (8 Epochs)")
    print("="*80)
    print("Fine-tuning Strategy Comparison:")
    print("-" * 50)

    for strategy, test_acc in test_accuracies.items():
        # Fix key mapping for results lookup
        if strategy == "ViT Full Fine-tune":
            key = "ViT_full_finetune"
        elif strategy == "ViT Frozen Backbone":
            key = "ViT_frozen_backbone"
        elif strategy == "ResNet50 Full Fine-tune":
            key = "ResNet50_full_finetune"
        elif strategy == "ResNet50 Frozen Backbone":
            key = "ResNet50_frozen_backbone"
        else:
            key = strategy

        val_acc = results[key][1]
        print(f"{strategy:25} | Val: {val_acc:5.2f}% | Test: {test_acc:5.2f}%")

    print(f"\nBest performing model: {best_model_name}")
    print(f"Best test accuracy: {test_accuracies[best_model_name]:.2f}%")

    # Key findings
    print(f"\n" + "="*50)
    print("KEY FINDINGS:")
    print("="*50)
    vit_full = test_accuracies.get("ViT Full Fine-tune", 0)
    vit_frozen = test_accuracies.get("ViT Frozen Backbone", 0)
    resnet_full = test_accuracies.get("ResNet50 Full Fine-tune", 0)
    resnet_frozen = test_accuracies.get("ResNet50 Frozen Backbone", 0)

    print(f"1. ViT vs ResNet (Full Fine-tuning): {vit_full:.2f}% vs {resnet_full:.2f}%")
    print(f"2. Full vs Frozen ViT: {vit_full:.2f}% vs {vit_frozen:.2f}%")
    print(f"3. Full vs Frozen ResNet: {resnet_full:.2f}% vs {resnet_frozen:.2f}%")
    print(f"4. Transfer learning effectiveness demonstrated!")
    print("5. Faster training with 8 epochs and batch size 64!")
    print("="*80)


if __name__ == "__main__":
    main()
```
